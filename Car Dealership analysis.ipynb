```python
df = dataset.drop(['id', 'VIN', 'year', 'region', 'country'], axis=1)
```

```python
from sklearn.preprocessing import OrdinalEncoder # For handling categorical values.
ord_enc = OrdinalEncoder()
df["manufacturer"] = ord_enc.fit_transform(df[["manufacturer"]])
df["model"] = ord_enc.fit_transform(df[["model"]])
df["condition"] = ord_enc.fit_transform(df[["condition"]])
df["cylinders"] = ord_enc.fit_transform(df[["cylinders"]])
df["fuel"] = ord_enc.fit_transform(df[["fuel"]])
df["title_status"] = ord_enc.fit_transform(df[["title_status"]])
df["transmission"] = ord_enc.fit_transform(df[["transmission"]])
df["drive"] = ord_enc.fit_transform(df[["drive"]])
df["size"] = ord_enc.fit_transform(df[["size"]])
df["type"] = ord_enc.fit_transform(df[["type"]])
df["paint_color"] = ord_enc.fit_transform(df[["paint_color"]])
df["state"] = ord_enc.fit_transform(df[["state"]])
```

```python
X = df.drop(['price'], axis=1)
```

```python
y = df['price']
```

```python
X
```

```python
y
```

```python
# Scaler is used to scale the data so that we can improve our model training.
from sklearn.preprocessing import StandardScaler # To Scale the data.
scaler = StandardScaler()
scaled_X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split # For splitting data into train and test data.
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import linear_model
from sklearn import svm
from sklearn import tree
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error
```

```python
model_1 = RandomForestRegressor(n_estimators = 400,min_samples_split = 4,min_samples_leaf = 1,max_features= 'sqrt',max_depth =None,bootstrap= False)
model_1.fit(X_train, y_train)
predict_1 = model_1.predict(X_test)
model_2= linear_model.Ridge()
model_2.fit(X_train, y_train)
predict_2 = model_2.predict(X_test)
model_3 =KNeighborsRegressor(10,weights='uniform')
model_3.fit(X_train, y_train)
predict_3 = model_3.predict(X_test)
model_4 = linear_model.BayesianRidge()
model_4.fit(X_train, y_train)
predict_4 = model_4.predict(X_test)
model_5 = tree.DecisionTreeRegressor(max_depth=1)
model_5.fit(X_train, y_train)
predict_5 = model_5.predict(X_test)
model_6= svm.SVR(C=1.0, epsilon=0.2)
model_6.fit(X_train, y_train)
predict_6 = model_6.predict(X_test)
model_7 = xgb.XGBRegressor()
model_7.fit(X_train, y_train)
predict_7 = model_7.predict(X_test)
```

```python
print(model_1.score(X_test, y_test))
print(model_2.score(X_test, y_test))
print(model_3.score(X_test, y_test))
print(model_4.score(X_test, y_test))
print(model_5.score(X_test, y_test))
print(model_6.score(X_test, y_test))
print(model_7.score(X_test, y_test))
```
